```python
import pandas as pd
huq=pd.read_sas('HUQ_H.XPT')
diq=pd.read_sas('DIQ_H.XPT')
df=huq.merge(diq, how='inner', on='SEQN')
```

```python
col_list=[]
for col in df.columns:
    if '175' in col:
        col_list.append(col)
df['175_num_factors']=df[col_list].count(axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['175_num_factors'].value_counts()
```

```python
df['HUQ051'][df['HUQ051']<1]=0
df['HUQ051'].value_counts()
```

```python
col_list=['HUQ010', 'HUQ020', 'HUQ030', 'HUQ041', 'HUQ051', 'HUQ061', 'DIQ010', 'DID040', '175_num_factors']
corr=df[col_list].corr()
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
plt.show()
```
